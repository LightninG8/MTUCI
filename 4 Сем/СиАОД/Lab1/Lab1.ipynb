Реализовать методы сортировки строк числовой матрицы в соответствии с

```python
import random
import time
import copy
import matplotlib.pyplot as plt
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.

# Функция сортировки выбором
def selection_sort(arg_arr):
  arr = copy.copy(arg_arr)

  for i in range(len(arr)):
    min_i = i
    
    for j in range(i + 1, len(arr)):
      if arr[j] < arr[min_i]:
        min_i = j
        
    arr[i], arr[min_i] = arr[min_i], arr[i] 
  
  return arr

def matrix_selection_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = selection_sort(matrix[i])
    
  return matrix


# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_selection_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
```

```python
# Сортировка вставкой.
# 1. Начинаем со второго элемента массива (первый элемент считается отсортированным).
# 2. Сравниваем текущий элемент с предыдущими и перемещаем его влево, пока не найдём правильное место.
# 3. Повторяем процесс для всех элементов массива, вставляя каждый новый элемент на правильное место в уже отсортированной части массива.
def insertion_sort(arg_arr):
  arr = copy.copy(arg_arr)
  
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key
  
  return arr

def matrix_insertion_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = insertion_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_insertion_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
```

```python
# Сортировка обменом. || Сортировка пузырьком.
def bubble_sort(arg_arr):
  arr = copy.copy(arg_arr)
  
  n = len(arr)
  for i in range(n - 1):
    for j in range(n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
  
  return arr

def matrix_bubble_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = bubble_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_bubble_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
```

```python
# Сортировка Шелла.
def shell_sort(arg_arr):
  arr = copy.copy(arg_arr)
  n = len(arr)
  gap = n // 2  # Начальный разрыв
  
  while gap > 0:
    for i in range(gap, n):
      temp = arr[i]
      j = i
      while j >= gap and arr[j - gap] > temp:
        arr[j] = arr[j - gap]
        j -= gap
      arr[j] = temp
    gap //= 2  # Уменьшаем разрыв
  
  return arr

def matrix_shell_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = shell_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_shell_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
```

```python
# Быстрая сортировка.
def quick_sort(arg_arr):
  arr = copy.copy(arg_arr)

  if len(arr) <= 1:
    return arr
  
  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]
  return quick_sort(left) + middle + quick_sort(right)

def matrix_quick_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = quick_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_quick_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
    
```

```python
# Турнирная сортировка.
def tournament_sort(arg_arr):
  arr = copy.copy(arg_arr)
  n = len(arr)
  tournament = arr[:]
  
  # Рекурсивная сортировка элементов турнира
  def sort_tournament(start, end):
    if end - start <= 1:
      return tournament[start:end]
    
    mid = (start + end) // 2
    left = sort_tournament(start, mid)
    right = sort_tournament(mid, end)
    
    # Слияние двух частей
    result = []
    while left and right:
      if left[0] < right[0]:
        result.append(left.pop(0))
      else:
        result.append(right.pop(0))
    
    result.extend(left if left else right)
    return result
  
  # Вызываем сортировку турнира на всем массиве
  return sort_tournament(0, n)

def matrix_tournament_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = tournament_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_tournament_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
    
```

```python
# Стандартная сортировка Python
def numeric_sort(arg_arr):
  arr = copy.copy(arg_arr)

  return sorted(arr)

def matrix_numeric_sort(arg_matrix):
  matrix = copy.copy(arg_matrix)
  
  for i in range(len(matrix)):
    matrix[i] = numeric_sort(matrix[i])
    
  return matrix

# Выполнение алгоритма
start_time = time.time()

sorted_matrix = matrix_numeric_sort(matrix)

print("\n--- {0} ms ---".format(round((time.time() - start_time)*1000)))

# Вывод результатов
print("\nИзначальная матрица:")
for row in matrix:
  print(row)

print("\nОтсортированная матрица:")
for sorted_row in sorted_matrix:
  print(sorted_row)
    
```
